```python
columns = ['rank', 'title', 'total ratings', 'installs', 'average rating',
       'growth (30 days)', 'growth (60 days)', 'price', 'category',
       '5 star ratings', '4 star ratings', '3 star ratings', '2 star ratings',
       '1 star ratings', 'paid']
messy_df = pd.DataFrame(data = [], columns=columns, index=[0])
clean_df= clean_columns(messy_df)
clean_df.columns.tolist()
```

```python
game_data = data.copy()
game_data.head()
```

```python
game_data.installs.value_counts()
```

```python
def to_million (inst):
    if inst == '500.0 k':
        return '0.5 M'
    elif inst == '100.0 k':
        return '0.1 M'
    else:
        return inst
```

```python
game_data['installs'] = game_data['installs'].apply(to_million)
game_data.installs.value_counts()
```

```python
game_data['installs'] = game_data['installs'].str.replace('M', '').str.strip().astype('float')
game_data['installs']
```

```python
game_data.installs.dtypes
```

```python
game_data = game_data.rename(columns={'installs': 'installs_in_millions'})
game_data
```

```python
game_data['5_star_percent'] = round(game_data['5_star_ratings']  / (game_data['5_star_ratings'] + game_data['4_star_ratings'] + game_data['3_star_ratings'] + game_data['2_star_ratings'] + game_data['1_star_ratings']) * 100, 2)
game_data['4_star_percent'] = round(game_data['4_star_ratings']  / (game_data['5_star_ratings'] + game_data['4_star_ratings'] + game_data['3_star_ratings'] + game_data['2_star_ratings'] + game_data['1_star_ratings']) * 100, 2)
game_data['3_star_percent'] = round(game_data['3_star_ratings']  / (game_data['5_star_ratings'] + game_data['4_star_ratings'] + game_data['3_star_ratings'] + game_data['2_star_ratings'] + game_data['1_star_ratings']) * 100, 2)
game_data['2_star_percent'] = round(game_data['2_star_ratings']  / (game_data['5_star_ratings'] + game_data['4_star_ratings'] + game_data['3_star_ratings'] + game_data['2_star_ratings'] + game_data['1_star_ratings']) * 100, 2)
game_data['1_star_percent'] = round(game_data['1_star_ratings']  / (game_data['5_star_ratings'] + game_data['4_star_ratings'] + game_data['3_star_ratings'] + game_data['2_star_ratings'] + game_data['1_star_ratings']) * 100, 2)
```
